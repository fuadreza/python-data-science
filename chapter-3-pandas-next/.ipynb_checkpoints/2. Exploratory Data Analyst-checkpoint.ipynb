# Display Visual

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
iris = pd.read_csv('iris.csv')
```

```python
iris.shape
```

```python
iris.head()
```

```python
iris.plot(x='sepal_length', y='sepal_width', kind='scatter')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
iris.plot(y='sepal_length', kind='box')
plt.ylabel('sepal_width (cm)')
```

```python
iris.plot(y='sepal_length', kind='hist')
plt.xlabel('sepal length (cm)')
```

```python
iris.plot(y='sepal_length', kind='hist', bins=30, range=(4,8))
plt.xlabel('sepal_length (cm)')
plt.show()
```

```python
iris.plot(y='sepal_length', kind='hist', bins=30, range=(4,8), cumulative=True)
plt.xlabel('sepal length (cm)')
plt.title('Cumulative distribution function (CDF)')
```

# Filtering

```python
iris.head()
```

```python
iris['species'].describe()
```

```python
iris['species'].value_counts()
```

```python
iris['species'].unique()
```

```python
index_setosa = iris['species'] == 'setosa'
index_versicolor = iris['species'] == 'versicolor'
index_virginica = iris['species'] == 'virginica'
```

```python
setosa = iris[index_setosa]
versicolor = iris[index_versicolor]
virginica = iris[index_virginica]
```
