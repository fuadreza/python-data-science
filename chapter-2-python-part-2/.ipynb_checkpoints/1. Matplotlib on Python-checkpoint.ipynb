# Matplotlib on Python

## Linear plot

```python
import matplotlib.pyplot as plt
age = [19, 20, 21, 22, 23, 24]
height_ratio = [167.011, 170.231, 172.657, 173.416, 174.313, 175.213]
plt.plot(age, height_ratio)
plt.show()
```

```python
plt.plot(age, height_ratio)
plt.xlabel('Age')
plt.ylabel('Height ratio')
plt.title('Class height ratio')
plt.yticks([164, 166, 168, 170, 172, 174, 176])
plt.show()
```

## Scatter plot

```python
plt.scatter (age, height_ratio)
plt.show()
```

## Hist plot

```python
score = [7.6, 7.8, 7.8, 7.9, 8.2, 8.5, 8.2, 7.7]
plt.hist(score, bins = 3)
plt.show()
```

```python
plt.hist(score, bins = 5)
plt.ylabel('Count')
plt.xlabel('Score')
plt.title('Score in Class B')
plt.show()
```
